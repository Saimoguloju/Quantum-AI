## 1. Two directions: 'Quantum AI' means two things

People use 'Quantum AI' for two distinct ideas — keep them separate:

1. **Quantum *for* AI** — using quantum computers to run or accelerate machine-learning tasks (the variational classifier in nb 03, quantum kernels, quantum-enhanced optimization).
2. **AI *for* Quantum** — using classical AI to *build better quantum computers*: discovering qubit materials, calibrating chips, designing error-correcting codes, and compiling circuits. (Microsoft used its **Discovery** agentic-AI to help develop the Majorana 2 chip — a concrete example.)

The most likely near-term winner is a **hybrid**: classical AI and quantum subroutines working together, not quantum replacing classical.

## 3. Where quantum advantage will likely appear first

Quantum computers won't be better at *everything* — they help where the problem has the right structure. The most promising areas, roughly in order of expected arrival:

| Domain | Why quantum helps | Status |
|--------|-------------------|--------|
| **Quantum chemistry / materials** | Simulating molecules *is* a quantum problem — nature is quantum | Most promising; early pilots in pharma & batteries |
| **Optimization** | Exploring huge solution spaces (logistics, finance, scheduling) | Active pilots; advantage not yet proven |
| **Machine learning** | High-dimensional feature spaces, quantum kernels, generative models | Promising but contested (see §5) |
| **Cryptography** | Shor's algorithm breaks RSA — *needs* fault tolerance | Years away, but driving 'post-quantum' security *now* |

Note: **quantum AI pilots are already live in 2026** in finance, pharma, and aerospace — but mostly as experiments, not yet delivering decisive real-world advantage.

## 4. The four families of Quantum ML

Modern QML (2026) clusters into four approaches — notebook 03 built the second one:

1. **Quantum kernels** — embed data into a vast `2ⁿ`-dimensional quantum space and measure similarities (overlaps) that may be hard to compute classically. There's even a *proven* advantage in one narrow case (the Liu–Arunachalam–Temme discrete-log kernel).
2. **Variational circuits (VQCs)** — trainable parameterised circuits, like the classifier you built (nb 03).
3. **Quantum generative models** — quantum versions of generative AI that learn to *produce* data/distributions.
4. **Quantum reinforcement learning** — agents whose policy or value function is computed by a quantum circuit.

The expected production pattern: **hybrid** models where classical deep-learning frameworks 'drop in' a quantum subroutine as one module — no need to redesign the whole AI stack.

## 5. The hard problems (be a realist, not a hype-merchant)

Two deep challenges define honest QML research today:

### Barren plateaus
As variational circuits get larger, their training gradients can shrink **exponentially** toward zero — the optimisation landscape becomes a vast, flat plain with no downhill direction to follow. Training stalls. 2026 mitigations include **problem-aware ansätze** (circuit designs tailored to the problem) and **local cost functions** that keep gradients measurable.

### The 'dequantization' paradox
Here's the sobering twist: many tricks that *avoid* barren plateaus (restricting the circuit to a small, well-behaved subspace) may also make the model **classically simulable** — i.e. a normal computer can do the same thing, erasing the quantum advantage. Several headline 'quantum advantage' claims have collapsed when researchers found efficient classical algorithms ('dequantization') that match them.

> **The honest summary:** provable, practical quantum advantage *for machine learning* is still **narrow and unproven** in general. This is exactly why the field is exciting — the foundational questions are still open.

## 6. A simple model of the hardware trend

Hardware progress has been roughly *exponential* in qubit count (a 'quantum Moore's law'), though raw qubit count matters far less than **error-corrected (logical) qubits**. Let's visualise the difference between scaling physical qubits and what actually counts — logical qubits.

````python
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2024, 2031)

# Illustrative ONLY — rough trend lines, not official figures.
physical_qubits = np.array([100, 130, 200, 400, 800, 1500, 3000])
# Logical qubits lag far behind because of error-correction overhead
logical_qubits  = np.array([0,   0,   1,   5,   20,  80,   200])

fig, ax = plt.subplots(figsize=(8, 4.5))
ax.plot(years, physical_qubits, "o-", label="Physical qubits (raw)")
ax.plot(years, logical_qubits,  "s-", label="Logical qubits (error-corrected)")
ax.set_yscale("log")
ax.set_xlabel("Year")
ax.set_ylabel("Qubits (log scale)")
ax.set_title("Illustrative quantum scaling: raw vs. error-corrected qubits")
ax.axhline(200, ls="--", color="gray", alpha=0.5)
ax.annotate("~200 logical qubits:\nIBM 'Starling' target (~2029)",
            xy=(2029, 200), xytext=(2024.3, 250), fontsize=9)
ax.legend()
plt.tight_layout()
plt.show()

print("NOTE: numbers above are illustrative to show the CONCEPT —")
print("logical qubits lag physical qubits because of error-correction overhead.")
````

## 7. How AI and quantum boost each other

The two technologies are increasingly intertwined:

- **AI improves quantum:** machine learning is used for qubit **calibration**, **error decoding** (interpreting QEC syndromes), **circuit compilation/optimisation**, and **materials discovery** (e.g. Microsoft Discovery for Majorana 2).
- **Quantum may improve AI:** potential speed-ups for sampling, optimisation, linear algebra, and richer feature spaces via quantum kernels.
- **Shared infrastructure:** quantum chips need the same cutting-edge classical compute (GPUs) that powers modern AI — for simulation, control, and error decoding in real time.

Expect the near future to be **co-evolution**, not competition.

## 8. Realistic outlook & how to get involved
