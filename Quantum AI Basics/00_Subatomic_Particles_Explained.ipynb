## 1. Zooming in: what is everything made of?

Pick up any object — your hand, a glass of water, the air. Keep cutting it into smaller and smaller pieces and you pass through a series of scales:

```
Everyday matter   (a drop of water)
      │  cut it down to the smallest piece that is still 'water'
      ▼
Molecule          (H₂O)                        ~0.3 nanometres
      │  molecules are made of atoms
      ▼
Atom              (hydrogen, oxygen…)          ~0.1 nanometres
      │  atoms have a tiny dense core + a cloud of electrons
      ▼
Subatomic         (protons, neutrons, electrons)
      │  protons & neutrons are themselves made of smaller things
      ▼
Fundamental       (quarks, electrons/leptons)  → no known smaller parts
```

**"Subatomic"** simply means *smaller than an atom*. Some subatomic particles (protons, neutrons) are **composite** — built from even smaller pieces. Others (electrons, quarks) are, as far as we know, **fundamental** — they have no internal parts at all.

> **Scale check:** if an atom were blown up to the size of a football stadium, the nucleus at its centre would be about the size of a pea. Atoms — and therefore you — are almost entirely empty space.

## 2. The atom: the first three subatomic particles

Every atom is built from just three particles:

| Particle | Where it lives | Electric charge | Relative mass | Role |
|----------|----------------|-----------------|---------------|------|
| **Proton** | In the nucleus | **+1** | ~1836× electron | Defines *which element* the atom is |
| **Neutron** | In the nucleus | **0** (neutral) | ~1839× electron | Adds mass, stabilises the nucleus |
| **Electron** | Orbiting cloud | **−1** | 1 (lightest) | Chemistry, electricity, bonding |

Key ideas:

- **The number of protons = the element.** 1 proton = hydrogen, 6 = carbon, 79 = gold. Change the proton count and you literally change the element.
- **Opposite charges attract.** Negative electrons are held to the positive nucleus by the **electromagnetic force**, like the Earth held to the Sun by gravity — except it's electric attraction, not gravity.
- **A neutral atom** has equal numbers of protons and electrons, so the charges cancel.
- **Electrons do the work** of chemistry and electronics. When electrons move, you get electric current; when atoms share or swap electrons, you get chemical bonds.

````python
# The actual masses (in kilograms) and charges of the three atomic particles.
# Notice how astonishingly tiny — and how much heavier the proton/neutron are than the electron.
particles = {
    #            mass (kg),        charge (Coulombs)
    "proton":   (1.67262192e-27,  +1.602176634e-19),
    "neutron":  (1.67492749e-27,   0.0),
    "electron": (9.1093837e-31,   -1.602176634e-19),
}

m_e = particles["electron"][0]
print(f"{'particle':10s} {'mass (kg)':>14s} {'charge (C)':>16s} {'mass vs electron':>18s}")
for name, (mass, charge) in particles.items():
    print(f"{name:10s} {mass:14.3e} {charge:16.3e} {mass/m_e:18.1f}")
````

## 3. Going deeper: the Standard Model

Protons and neutrons are **not** fundamental — each is made of three smaller particles called **quarks**, glued together by the **strong force**.

- A **proton** = two *up* quarks + one *down* quark  (charges: +2/3 +2/3 −1/3 = **+1** ✔)
- A **neutron** = one *up* quark + two *down* quarks (charges: +2/3 −1/3 −1/3 = **0** ✔)

All known matter is described by the **Standard Model of particle physics** — our best, exhaustively tested theory of the fundamental particles. The matter particles (called **fermions**) come in two families:

### Quarks (feel the strong force, live inside protons/neutrons)
| Generation | 'Up-type' (+2/3) | 'Down-type' (−1/3) |
|------------|------------------|---------------------|
| 1 (everyday matter) | up (u) | down (d) |
| 2 (heavier) | charm (c) | strange (s) |
| 3 (heaviest) | top (t) | bottom (b) |

### Leptons (do NOT feel the strong force)
| Generation | Charged (−1) | Neutral (neutrino) |
|------------|--------------|--------------------|
| 1 | electron (e) | electron-neutrino (νₑ) |
| 2 | muon (μ) | muon-neutrino (ν_μ) |
| 3 | tau (τ) | tau-neutrino (ν_τ) |

Almost everything you ever see is made from just the **first generation**: up quarks, down quarks, and electrons. The heavier generations are unstable and decay quickly — we mostly create them in particle accelerators or see them in cosmic rays.

**Neutrinos** are ghostly: nearly massless, no electric charge, and they barely interact. Trillions stream through your body every second from the Sun without touching a thing.

## 4. Force carriers (bosons) and the four forces

Matter particles don't just sit there — they **interact**. In the Standard Model, forces are transmitted by exchanging another class of particles called **bosons** (the 'force carriers'). Think of two ice skaters throwing a ball back and forth: the act of exchanging the ball pushes them around. Particles 'throw' bosons to push and pull on each other.

| Force | Carrier boson | Strength | Range | What it does |
|-------|---------------|----------|-------|--------------|
| **Strong** | gluon (g) | strongest | tiny (nucleus) | Binds quarks into protons/neutrons; holds the nucleus together |
| **Electromagnetic** | photon (γ) | strong | infinite | Light, electricity, magnetism, chemistry |
| **Weak** | W and Z bosons | weak | tiny | Radioactive decay; powers the Sun's fusion |
| **Gravity** | (graviton — hypothetical) | weakest | infinite | Attracts all mass/energy (not yet in the Standard Model) |

Plus one very special particle:

- The **Higgs boson** — discovered in 2012 at CERN. The Higgs *field* is what gives many particles their **mass**. Particles that interact strongly with it (like the top quark) are heavy; those that don't (like the photon) are massless and travel at light speed.

> **The photon** is the star of quantum technology: it is a single particle ('quantum') of light. Lasers, fibre-optic internet, and many quantum computers all manipulate individual photons.

## 5. How do subatomic particles actually *work*?

At this scale, intuition from everyday objects breaks down. Particles obey **quantum mechanics**. The most important properties and behaviours:

### Properties every particle has
- **Mass** — how much it resists being accelerated (given by the Higgs field).
- **Electric charge** — whether and how it feels the electromagnetic force.
- **Spin** — an intrinsic 'angular momentum', a bit like the particle carries a built-in tiny magnet. Spin is *quantised*: an electron's spin can only be 'up' or 'down', nothing in between. **This two-state spin is one physical way to build a qubit.**

### The strange quantum behaviours
- **Wave–particle duality.** A particle is *also* a wave. An electron can interfere with itself like a ripple on a pond, yet arrive at a detector as a single dot. It behaves as both, depending on how you look.
- **Superposition.** Before measurement, a particle can be in a blend of possibilities at once (e.g. spin up *and* down). (This is the core idea behind notebook 01.)
- **Quantisation.** Many quantities — energy levels in an atom, spin, charge — come only in discrete 'steps', not a smooth continuum. The word *quantum* literally means a discrete amount.
- **Uncertainty principle.** You cannot know both a particle's exact position and exact momentum at the same time. The more precisely you pin down one, the fuzzier the other becomes.
- **Quantum fields.** The deepest modern view: each particle is a localised 'ripple' or excitation of an underlying field that fills all of space. An electron is a ripple in the electron field; a photon is a ripple in the electromagnetic field.

### A concrete number: wave–particle duality
Every particle has a **de Broglie wavelength** $\lambda = h / (m v)$, where $h$ is Planck's constant. Heavy/fast things have wavelengths far too tiny to notice (which is why a thrown baseball looks like a particle, not a wave). Light, slow things like electrons have wavelengths big enough to matter.

````python
h = 6.62607015e-34          # Planck constant (J·s)
m_e = 9.1093837e-31         # electron mass (kg)

def de_broglie_wavelength(mass, velocity):
    """Quantum wavelength of a moving object, in metres."""
    return h / (mass * velocity)

# An electron moving at 1% the speed of light
v_electron = 0.01 * 3e8
lam_e = de_broglie_wavelength(m_e, v_electron)

# A 0.145 kg baseball thrown at 40 m/s
lam_ball = de_broglie_wavelength(0.145, 40)

print(f"Electron wavelength: {lam_e:.3e} m  (~size of an atom -> wave effects are HUGE)")
print(f"Baseball wavelength: {lam_ball:.3e} m  (~10^-34 m -> utterly unmeasurable)")
print("\nThis is WHY tiny particles act 'quantum' and big objects act 'classical'.")
````

````python
# Light itself is made of particles: photons. A photon's energy depends only on its frequency:
#     E = h * f
# This is why blue/UV light (high frequency) is more energetic — and more damaging — than red light.
def photon_energy(frequency_hz):
    return h * frequency_hz

for color, freq in [("red light", 4.3e14), ("green light", 5.6e14), ("ultraviolet", 1.0e15)]:
    E = photon_energy(freq)
    print(f"{color:12s}: each photon carries {E:.3e} joules")
````

## 6. Antimatter

Every matter particle has a mirror-image **antiparticle** with the *opposite charge* but the *same mass*. The electron's antiparticle is the **positron** (charge +1).

When a particle meets its antiparticle they **annihilate**, converting their entire mass into energy according to Einstein's $E = mc^2$. This isn't science fiction — it's used every day in hospitals (see PET scans below).

One of the great unsolved puzzles of physics: the Big Bang should have made equal amounts of matter and antimatter, which would have annihilated completely. Yet here we are. Why matter won is still an open question.

## 7. How do we study things we can't see?

Subatomic particles are far too small to see with any microscope. Instead, physicists:

- **Smash them together** in *particle accelerators* like the Large Hadron Collider (LHC) at CERN — a 27 km ring that accelerates protons to 99.9999991% the speed of light and collides them. The energy of the collision creates new, heavier particles (this is how the Higgs boson was found).
- **Detect the debris** with giant *detectors* that track the paths, charges, and energies of particles flying out of a collision.
- **Catch rare particles** in deep underground labs (shielded from cosmic rays) — for example, enormous tanks of fluid waiting for a neutrino to interact.
- **Watch the cosmos**, since stars, supernovae, and cosmic rays are natural particle factories.

From these experiments we measure each particle's mass, charge, spin, and lifetime, and test the Standard Model to extraordinary precision.

## 8. What are subatomic particles *used for*?

Understanding and controlling particles isn't just abstract — it underpins a huge fraction of modern technology.

### Electronics & computing
- **Electrons** are the workhorses of all electronics. Every transistor in your phone or laptop works by controlling the flow of electrons. Modern chips rely on *quantum* effects (electron tunnelling) at nanometre scales.

### Light & communication
- **Photons** carry the internet through fibre-optic cables, power lasers, and enable cameras and solar panels (where photons knock electrons loose to make electricity — the *photoelectric effect*).

### Medicine
- **PET scans** (Positron Emission Tomography) use *antimatter*: a tracer emits positrons that annihilate with electrons in your body, producing photons the scanner detects to image tumours.
- **Radiotherapy** uses beams of particles or high-energy photons to destroy cancer cells.
- **MRI** exploits the *spin* of protons in your body's water molecules.

### Energy
- **Nuclear power & the Sun** both come from the nucleus: fission splits heavy nuclei; fusion (in the Sun and future reactors) fuses light ones — both governed by the strong and weak forces between subatomic particles.

### Materials & dating
- **Radioactive decay** (a weak-force process) lets us carbon-date fossils and archaeological finds.

### Quantum computing — the reason this notebook is here
- A **qubit** is built from a controllable two-state quantum property of a particle. Common choices:
  - the **spin** of an electron or atomic nucleus (up / down),
  - the **polarisation or path** of a single **photon**,
  - the energy levels of an electron in a 'trapped ion' or a superconducting circuit.
- The superposition and entanglement of these particles (notebooks 01–03) are exactly what give quantum computers their power.
