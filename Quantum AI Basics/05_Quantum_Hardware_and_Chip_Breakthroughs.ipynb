## 1. Why building a quantum computer is so hard

Qubits are extraordinarily **fragile**. The slightest disturbance — heat, vibration, stray electromagnetic fields — destroys their delicate superposition. Two key problems:

- **Decoherence:** a qubit holds its quantum state only for a tiny fraction of a second before the environment scrambles it. This is the qubit's *coherence time*.
- **Errors / noise:** gates aren't perfect; every operation introduces a small error. Run enough gates and the errors swamp the answer.

The fix is **Quantum Error Correction (QEC)**: spread the information of one reliable *logical qubit* across many noisy *physical qubits*, so errors can be detected and corrected. The catch — it can take **hundreds or thousands of physical qubits to make one good logical qubit**. This overhead is the central engineering battle of the entire field.

Most quantum chips are kept colder than outer space (near **absolute zero, ~15 millikelvin**) inside dilution refrigerators to keep decoherence at bay.

## 2. The main types of qubit hardware

There is no single agreed-upon way to build a qubit. The leading approaches each trade off speed, stability, and scalability:

| Technology | How the qubit is made | Who | Pros | Cons |
|------------|-----------------------|-----|------|------|
| **Superconducting** | Tiny supercooled circuits with quantised energy levels | Google, IBM | Fast gates, mature fabrication | Short coherence, needs near-0 K |
| **Trapped ions** | Individual atoms held by electromagnetic fields, controlled by lasers | IonQ, Quantinuum | Very high fidelity, long coherence | Slower gates |
| **Photonic** | Single particles of light (photons) | PsiQuantum, Xanadu | Works at room temperature, networkable | Hard to make photons interact |
| **Neutral atoms** | Atoms held in 'optical tweezers' | QuEra, Pasqal | Scales to many qubits, flexible | Newer, still maturing |
| **Topological** | Exotic 'Majorana' quasiparticles | Microsoft | *In principle* error-resistant by design | Unproven physics, very early |

The notebooks 01–03 ran on a **superconducting-style** simulator (Qiskit's default).

## 3. Google — the Willow chip 🟢

Google's quantum processor **Willow** (superconducting) has driven two headline breakthroughs.

### Breakthrough 1 — Below-threshold error correction (unveiled Dec 2024)
Willow (a **65-qubit** device with higher-coherence qubits) was the first chip to demonstrate that **adding more qubits can make the logical qubit *better*, not worse**. As they scaled up the error-correcting grid, the error rate *fell exponentially*. This 'below-threshold' result experimentally validates the error-correction theory Peter Shor proposed in 1995 — a crucial milestone on the road to fault tolerance.

### Breakthrough 2 — Verifiable quantum advantage (Oct 2025)
Google ran an algorithm called **Quantum Echoes** (based on an *out-of-time-order correlator*, OTOC) and reported the first **verifiable quantum advantage**: Willow completed the task about **13,000× faster** than the best classical supercomputers, and — crucially — the result is *verifiable* and repeatable, addressing a key criticism of earlier 'quantum supremacy' claims. In Dec 2025 Google began opening Willow access to outside (e.g. UK) researchers to hunt for practical applications.

> **Why 'verifiable' matters:** earlier supremacy demos produced answers that were hard to independently check. A verifiable advantage is far more convincing evidence of real quantum usefulness.

## 4. Microsoft — the Majorana topological bet 🟦

Microsoft took a radically different, higher-risk path: **topological qubits**.

### Majorana 1 (Feb 2025)
Microsoft unveiled **Majorana 1**, which it called the world's first quantum processor powered by a **Topological Core**. The idea: engineer exotic **Majorana Zero Modes** at the ends of *topological superconducting nanowires* (built from **indium arsenide + aluminium**, cooled near absolute zero). Because the quantum information is stored *non-locally* in the topology, such a qubit would be **intrinsically protected from noise** — error resistance built into the physics rather than bolted on via QEC. Microsoft's stated ambition: scale to **a million qubits on a single chip**.

### Majorana 2 (June 2026)
At Build 2026, Microsoft unveiled **Majorana 2**, claiming a **~1,000× improvement in qubit stability** over Majorana 1, with a **mean qubit lifetime of ~20 seconds** (some instances up to a minute). It uses a new materials stack (**lead** instead of aluminium) and was developed with help from **Microsoft Discovery**, an agentic-AI system for materials research. Microsoft also pulled its timeline for a scalable quantum computer **forward from 2033 to 2029**.

### The controversy
Topological qubits remain **scientifically contested**. Whether Microsoft has truly created and controlled Majorana particles — or is measuring some other effect — is one of the most debated questions in condensed-matter physics. Many researchers remain sceptical. This is a genuine *high-risk, high-reward* bet: if it works, it could leapfrog everyone; if the physics doesn't hold up, it's back to the drawing board.

## 5. IBM — the roadmap to fault tolerance ⬛

IBM has published the most detailed public **roadmap**, emphasising modular, scalable systems and a shift from surface codes to **quantum LDPC (qLDPC) codes** — which cut the physical-qubit overhead by up to **~90%**.

| Processor / step | Target | Role |
|------------------|--------|------|
| **Nighthawk** | end of 2025 | 120 qubits, 218 tunable couplers; aims for **quantum advantage by end of 2026** |
| **Loon** | 2025 | Demonstrates architecture for qLDPC error-correcting codes |
| **Kookaburra** | 2026 | First **fault-tolerant module** (logic + memory) |
| **Cockatoo** | 2027 | Entanglement *between* modules |
| **Starling** | 2028–2029 | First **large-scale fault-tolerant** machine: ~**200 logical qubits**, **100 million gates** |

IBM's two headline targets: **quantum advantage by 2026**, and a **large-scale fault-tolerant quantum computer by 2029**.

> **Surface codes vs. qLDPC:** both are error-correcting schemes. qLDPC codes need far fewer physical qubits per logical qubit, which is why IBM is betting on them to make fault tolerance practical.

## 7. NISQ vs. fault-tolerant — where we *really* are

Two important eras to keep straight:

- **NISQ — Noisy Intermediate-Scale Quantum (today).** Chips with tens to a few hundred qubits, *no* full error correction. Useful for research, experiments, and the kind of variational QML in notebook 03 — but noise limits how deep circuits can go.
- **Fault-tolerant (the goal).** Enough error correction that arbitrarily long computations run reliably. This is what unlocks the famous applications (breaking RSA encryption, simulating complex molecules for drug discovery and new materials). IBM targets ~2029; Microsoft also now says ~2029; most experts expect *useful* fault tolerance somewhere in the late 2020s to 2030s.

**Bottom line (June 2026):** we are in the late-NISQ era, with the *first* convincing demonstrations of quantum error correction and verifiable advantage now in hand — but a general-purpose, fault-tolerant quantum computer does **not** yet exist. The next 3–5 years are about scaling error correction from a lab milestone into real machines.
